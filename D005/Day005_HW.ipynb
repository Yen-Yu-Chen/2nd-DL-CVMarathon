## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests 

rs = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true') #想要爬資料的目標網址
response = rs.text #模擬發送請求的動作
len(response) 
response
```

```python
import json
data = json.loads(response) #利用json解析網頁內容
data
```

```python

len(data) #一次回傳幾筆資料
```

```python
for d in data[0]: #每一筆資料包含哪些欄位
    print(d)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

for d in data: #每一筆資料包含哪些欄位
    print(d["title"], d["createdAt"],d["commentCount"],d["likeCount"])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

for d in data: 
    d_comment = d["commentCount"]
    print(d_comment)
```

```python
sum_c = 0
for i in d_comment:
    sum_c = sum_c + i
    print(d_comment)
```
